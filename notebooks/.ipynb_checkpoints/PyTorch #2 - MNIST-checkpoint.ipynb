## Data Structure

```
[offset] [type]          [value]          [description]
0000     32 bit integer  0x00000803(2051) magic number
0004     32 bit integer  60000            number of images
0008     32 bit integer  28               number of rows
0012     32 bit integer  28               number of columns
0016     unsigned byte   ??               pixel
0017     unsigned byte   ??               pixel
........
xxxx     unsigned byte   ??               pixel
```

````python
import struct

def read_struct(fp, fmt):
    return struct.unpack(fmt, fp.read(struct.calcsize(fmt)))

with open('../data/t10k-images-idx3-ubyte', 'rb') as fp:
#     header = fp.read(16)
#     magic, num_images, num_rows, num_cols = struct.unpack(">iiii", header)
    magic, num_images, num_rows, num_cols = read_struct(fp, ">iiii")
    assert magic == 2051
    
    print(magic, num_images, num_rows, num_cols)
    
    image = read_struct(fp, f"<{num_rows * num_cols}B")
    image = np.array(image).reshape((num_rows, num_cols))
    plt.imshow(image)
````

````python
HEADER_FMT = ">iiii"

class MNISTDataset(Dataset):
    
    def __init__(self, image_file, label_file=None):
        # Read header
        self.image_fp = open(image_file, 'rb')
        if label_file:
            self.label_fp = open(label_file, 'r')
            
        magic, self.num_images, self.num_rows, self.num_cols = read_struct(self.image_fp, ">iiii")
        self.offset = struct.calcsize(HEADER_FMT)
        
    def __del__(self):
        self.image_fp.close()
        
    def __len__(self):
        return self.num_images
    
    def __getitem__(self, idx):
        if idx < len(self):
            fpi = idx * self.num_rows * self.num_cols + self.offset
#             print(fpi)
            self.image_fp.seek(fpi, 0)
            struct = read_struct(self.image_fp, f"<{self.num_rows * self.num_cols}B")
#             self.image_fp.seek(16)
#             image = np.array()
            image = torch.Tensor(struct)
#             image = np.array(image).reshape((num_rows, num_cols))
            self.image_fp.seek(16)
            return image
        else:
            raise ValueError("Out of bounds")
            
ds = MNISTDataset('../data/t10k-images-idx3-ubyte')
# print(ds[0])
X = [ds[0], ds[1], ds[12], ds[100]]
images = np.vstack([np.reshape(x, (ds.num_rows, ds.num_cols)) for x in X])
plt.imshow(images)
````

````python
import struct

def read_struct(fp, fmt):
    return struct.unpack(fmt, fp.read(struct.calcsize(fmt)))

with open('../data/t10k-labels-idx3-ubyte', 'rb') as fp:
#     header = fp.read(16)
#     magic, num_images, num_rows, num_cols = struct.unpack(">iiii", header)
    magic, num_labels = read_struct(fp, ">ii")
    assert magic == 2049
    
    print(magic, num_labels)
    read_struct(fp, f">{num_labels}B")
    print()
````

````python
in_size = ds.num_rows*ds.num_cols
num_classes = 10
print(in_size)
model = nn.Sequential(
    nn.Linear(in_size, 1024),
    nn.Linear(1024, num_classes)
)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
````

````python
num_epochs = 1

for ep in range(num_epochs):
    total_loss = 0
    dl = DataLoader(ds, shuffle=True, batch_size=128, num_workers=2)
    for i, data in enumerate(dl):
        X, y = data
    
        y_pred = model(X)
        loss = optimizer(y, y_pred)
        loss.backward()
        optimizer.zero_grad()
        new_loss = loss.item()
        total_loss += new_loss
        optimizer.step()
        
        if i % 2000:
            print("total_loss=", total_loss)
````
